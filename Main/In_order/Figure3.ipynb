```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='serif', size=15)

# Define the x values
x = np.linspace(-1, 1, 400)
x
# Define the functions
y1 = -x
y2 = -np.sin(np.pi/2 * x)
y3 = -np.sin(np.pi/2 * x)**3
y4 = -np.sin(np.pi/2 * x)**5

# Plot the functions
plt.figure(figsize=(7, 7))

plt.plot(x, y1, label='x', linestyle='--',linewidth=3)
plt.plot(x, y2, label=r'1', linestyle='-.',linewidth=3)
plt.plot(x, y3, label=r'3', linestyle=':',linewidth=3)
plt.plot(x, y4, label=r'5',linewidth=3)
plt.fill_between(x,y4,0,alpha=0.3)

# Add titles and labels

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle='--')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.savefig('plot.svg')

# Show the plot
plt.show()

```
